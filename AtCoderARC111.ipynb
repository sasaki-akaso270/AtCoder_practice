```
#AtCoder A問題
import math
def pow1(x, n):
    if n == 0: return 1
    value = pow1(x, int(n / 2))
    value *= value
    if n % 2 == 1: value *= x
    return value
n, m = map(int , input().split())
res = pow1(10,n) // m
print(res%m)
```

```
#AtCoder B問題
import pandas as pd
import collections

n = int(input())
nlist = []
for i in range(n):
  a1 , a2 = map(int, input().split())
  nlist.append([a1,a2])

n1lst = list(map(lambda x: x[0], nlist))
n2lst = list(map(lambda x: x[1], nlist))

tmp = list(set(n1lst).union(n2lst))

d = collections.Counter(n1lst + n2lst)
val = [1 for k, v in d.items() if v == 1]

if len(tmp) < n:
  print(len(tmp))
else:
  print(n)

```
